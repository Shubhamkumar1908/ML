```python
import matplotlib.pyplot as plt

categories = ['Math', 'science', 'english', 'history']
scores = [89,92,80,70]

plt.bar(categories,scores,color = 'skyblue')
plt.title('student scores')
plt.xlabel('subject')
plt.ylabel('scores')
plt.grid(axis='y', linestyle='--')
plt.show()
```

```python
import matplotlib.pyplot as plt

categories = ['Q1','Q2','Q3','Q4']
product_A = [40,60,80,70]
product_B = [30,50,60,90]

plt.figure(figsize=(8, 6))
plt.bar(categories, product_A, label='Product A', color='skyblue')
plt.bar(categories, product_B, label='Product B', color='lightcoral', bottom=product_A)

plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.title('Quarterly Sales of Product A and Product B')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_1 = np.linspace(0,5,10)
y_1 = x_1**2

fig_1 = plt.figure(figsize=(10, 6),dpi=80)
axes_1 = fig_1.add_subplot(1,1,1)
axes_1.set_xlabel('Days')
axes_1.set_ylabel('Day Squard')
axes_1.set_title('Days squard char')
axes_1.plot(x_1,y_1)

axes_1.plot(x_1,y_1,label='x / x^2')
axes_1.plot(y_1,x_1,label='x^2 / x')

axes_1.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.linspace(0, 10, 100)

y = 2 * x + 3

plt.plot(x, y)

plt.title('Line Plot of y = 2x + 3')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)

y = x**2

plt.plot(x, y, 'r--', linewidth=2, marker='o')

plt.title('Plot of y = x')
plt.xlabel('x')
plt.ylabel('y')

plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)

y1 = x**2
y2 = x**3

plt.plot(x, y1, 'r--', label='y = x')
plt.plot(x, y2, 'b-', label='y = x')

plt.title('Plots of y = x and y = x')
plt.xlabel('x')
plt.ylabel('y')

plt.legend()

plt.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2 * np.pi, 100)

y_sin = np.sin(x)
y_cos = np.cos(x)

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(x, y_sin, 'b-')
plt.title('sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.grid(True)

plt.subplot(1, 2, 2)
plt.plot(x, y_cos, 'r--')
plt.title('cos(x)')
plt.xlabel('x')
plt.ylabel('cos(x)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x, y, color='green', s=50)

plt.title('Scatter Plot of Random Values')
plt.xlabel('X-values')
plt.ylabel('Y-values')

plt.show()
```
